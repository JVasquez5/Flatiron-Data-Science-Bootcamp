### Box Office Mojo Web Scrapping
The Box Office Mojo(BOM) website was extremely user-friendly and provided exactly the right pieces of information in order to provide actionable insight regarding entrance into the movie business. 

```python
import requests #To get url
from bs4 import BeautifulSoup as bs #To perform web scraping
```

```python
r = requests.get("https://www.boxofficemojo.com/chart/ww_top_lifetime_gross/") #For top 200 lifetime grossing movies
soup1 = bs(r.content, "html.parser")
```

```python
print(soup1.prettify()) #To view info slightly easier
```

While the content that was provided after the webscrap, contained relevant information; it was still dense, extremely cluttered, and not aesthetically pleasing. As such, additional steps were required in order to solely extract the relevant pieces of information. Thus, upon further inspect within the BOM website I was able to pinpoint, through the utilization of the "Inspect Element" feature, the 'tr' & 'td' tags contained all the specific elements needed for my table. Furthermore, also embbeded within the content scrapped were the IMDB indentification numbers. Therefore, extracting the information was ideal as it served a dual purpose of identifying each individual movie, as well as providing a easy way to merge future datasets with corresponding identification.

```python
movie_list1 = []
for record in soup1.findAll("tr"): 
    row = [x.text for x in record.findAll("td")]
    for a in record.findAll('a', href=True):
        if a['href'].startswith('/title/'):
            row.append(a['href'][7:16])
    movie_list1.append(row)
print(movie_list1)
```
